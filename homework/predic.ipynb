```python
import pandas as pd
import pickle
import gzip

# Ruta del modelo
MODEL_PATH = "/Users/dmc/Trabajo 2025/MOCA /repository/MOCA/file/models/modeloCBC.pkl.gz"

# Variables categóricas y numéricas
categorical_features = [
    'Gender', 'EducationLevel', 'Smoking', 'FamilyHistoryParkinsons',
    'TraumaticBrainInjury', 'Hypertension', 'Diabetes', 'Depression',
    'Stroke', 'Tremor', 'Rigidity', 'Bradykinesia', 'PosturalInstability',
    'SpeechProblems', 'SleepDisorders', 'Constipation', 'Diagnosis'  # Diagnóstico es target
]

numerical_features = [
    'Age', 'BMI', 'AlcoholConsumption', 'PhysicalActivity', 'DietQuality',
    'SleepQuality', 'SystolicBP', 'DiastolicBP', 'CholesterolTotal',
    'CholesterolLDL', 'CholesterolHDL', 'CholesterolTriglycerides', 'MoCA'
]

# Función para cargar el modelo
def cargar_modelo(ruta):
    with gzip.open(ruta, 'rb') as f:
        modelo = pickle.load(f)
    return modelo

# Función para predecir
def predecir_diagnostico(datos_entrada):
    modelo = cargar_modelo(MODEL_PATH)

    # Crear DataFrame de entrada
    df = pd.DataFrame([datos_entrada])

    # Eliminar 'Diagnosis' si está presente (target)
    if 'Diagnosis' in df.columns:
        df = df.drop(columns=['Diagnosis'])

    # Predecir
    prediccion = modelo.predict(df)
    return prediccion[0]

# Ejemplo de uso con datos simulados
datos_simulados = {
    'Gender': 1,
    'EducationLevel': 2,
    'Smoking': 0,
    'FamilyHistoryParkinsons': 1,
    'TraumaticBrainInjury': 0,
    'Hypertension': 1,
    'Diabetes': 0,
    'Depression': 1,
    'Stroke': 0,
    'Tremor': 1,
    'Rigidity': 1,
    'Bradykinesia': 1,
    'PosturalInstability': 0,
    'SpeechProblems': 1,
    'SleepDisorders': 1,
    'Constipation': 1,
    'Age': 74,
    'BMI': 26.4,
    'AlcoholConsumption': 4,
    'PhysicalActivity': 2,
    'DietQuality': 7,
    'SleepQuality': 6,
    'SystolicBP': 140,
    'DiastolicBP': 85,
    'CholesterolTotal': 220,
    'CholesterolLDL': 135,
    'CholesterolHDL': 45,
    'CholesterolTriglycerides': 160,
    'MoCA': 22
}


# Ejecutar la predicción
resultado = predecir_diagnostico(datos_simulados)
print("Diagnóstico predicho:", resultado)
```
